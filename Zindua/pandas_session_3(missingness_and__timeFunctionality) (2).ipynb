**- Let's be carefull on such methods**

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with a missing value in the last row
data2 = {'Column1': [1, 2, 3, np.nan],
        'Column2': ['A', 'B', 'C', np.nan]}

df2 = pd.DataFrame(data2)

# Perform backfill along rows (axis=0)
df_backfilled = df2.bfill()  #.fillna(method='bfill')

# Display the original and backfilled DataFrames
print("Original DataFrame:")
print(df2)
print("\nDataFrame after backfill:")
print(df_backfilled)
```
