```python
plt.figure(figsize=(10,5))
plt.plot(data['date'],data['COTTON'],label = 'cotton')
plt.plot(data['date'],data['COTTON'])
plt.plot(data['date'],data['WTI'],label='wti')
plt.plot(data['date'],data['NATURAL_GAS'],label='gas')
plt.title("Cotton & WTI Price Over Time")
plt.xlabel("Date")
plt.ylabel("Cotton Price")
plt.legend()
plt.show()
```

```python
#creating sublplots
fig,ax = plt.subplots(1,2,figsize =(16,4))
ax[0].plot(data['date'],data['WTI'],color = 'magenta')
ax[1].plot(data['date'],data['NATURAL_GAS'],color ='blue')
ax[0].set_title("WTI Price")
ax[1].set_title("Natural Gas Price")
ax[0].set_xlabel("date")
ax[1].set_xlabel("DATE")
ax[0].set_ylabel("WTI price")
ax[1].set_ylabel("Gas Price")
plt.savefig("demo.png")
plt.show()
```

```python
fig,ax =plt.subplots(2,3,figsize = (16,10))
ax[0,0].plot(data['date'],data['SUGAR'],color = 'magenta')
ax[0,1].plot(data['date'],data['WHEAT'],color ='blue')
ax[0,2].plot(data['date'],data['ALUMINUM'],color='yellow')
ax[1,0].plot(data['date'],data['COFFEE'],color = 'green')
ax[1,1].plot(data['date'],data['CORN'],color ='red')
ax[1,2].plot(data['date'],data['BRENT'],color ='orange')
ax[0,0].set_title("SUGAR")
ax[0,1].set_title("WHEAT")
ax[0,2].set_title("ALUMINUM")
ax[1,0].set_title("COFFEE")
ax[1,1].set_title("CORN")
ax[0,2].set_title("BRENT")

ax[0,0].set_xlabel("date")
ax[0,1].set_xlabel("date")
ax[0,2].set_xlabel("date")
ax[1,0].set_xlabel("date")
ax[1,1].set_xlabel("date")
ax[1,2].set_xlabel("date")

ax[0,0].set_ylabel("Sugar price")
ax[0,1].set_ylabel("Wheat Price")
ax[0,2].set_ylabel("Aluminum Price")
ax[1,0].set_ylabel("Coffee price")
ax[1,1].set_ylabel("Corn Price")
ax[1,2].set_ylabel("Brent Price")

plt.show()
```
