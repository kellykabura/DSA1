```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/quantium_merged_data.csv")
```

```python
data.info()
```

**Distribution Plots**

```python
data.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data=data,x='TOT_SALES',bins=20)
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
data['TOT_SALES'].describe()
```

**Categories ; PREMIUM CUSTOMER**

```python
sns.kdeplot(data=data,x='TOT_SALES',hue='PREMIUM_CUSTOMER',fill=True)
plt.show()
```

**Boxplot**

```python
sns.boxplot(data=data,x='TOT_SALES')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=data,y='LIFESTAGE',x='PACKET_SIZE',hue='PREMIUM_CUSTOMER')
plt.show()
```
