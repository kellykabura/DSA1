**Introduction to Data Visualization**

Data visualization is a powerful tool in the field of data analysis and interpretation. As humans, our brains are wired to process visual information quickly and efficiently. Therefore, representing data visually allows us to uncover patterns, trends, and insights that might be challenging to grasp from raw numbers alone. In this context, the use of plots and charts becomes indispensable.

**Why Do We Plot?**

**Pattern Recognition:** Plots provide a visual summary of data, making it easier to identify patterns and trends. Patterns that might go unnoticed in tables of numbers become strikingly clear when presented graphically.

**Communication:** Visualizations are a universal language that transcends barriers. They facilitate effective communication of complex information to a diverse audience, including stakeholders, decision-makers, and the general public.

**Comparison:** Plots enable us to compare different variables, groups, or time periods, helping us draw meaningful comparisons and make informed decisions.

**Storytelling:** A well-crafted plot can tell a story. It can guide the audience through the narrative of your data, emphasizing key points and insights.

**Common Types of Plots and Their Insights**

1. Line Plots:
Use: Show trends over a continuous variable (e.g., time).
Insights: Identify upward or downward trends, seasonality, or cyclical patterns.
2. Bar Plots:
Use: Compare categories or groups.
Insights: Highlight differences in magnitude between categories, visualize distributions, and show rankings.
3. Histograms:
Use: Display the distribution of a single variable.
Insights: Identify the central tendency, spread, and shape of the distribution.
4. Scatter Plots:
Use: Explore relationships between two continuous variables.
Insights: Identify correlations, clusters, outliers, or trends.
5. Pie Charts:
Use: Show the proportion of parts to a whole.
Insights: Illustrate the contribution of each category to the total.
6. Heatmaps:
Use: Visualize relationships in a matrix format.
Insights: Identify patterns, correlations, or trends in a multidimensional dataset.
7. Box Plots:
Use: Display the distribution of a continuous variable, highlighting central tendency and variability.
Insights: Identify outliers, compare distributions, and understand the spread of the data.

```python
x = data['date']
y = data['COTTON']
y1 = data['WTI']
y2 = data['BRENT']

plt.figure(figsize=(14,4))
plt.plot(x,y,label='Cotton',color='green',linestyle=":",linewidth=0.5)
plt.plot(x,y1,label='WTI',color = 'maroon',linestyle="--",linewidth=1.5)
plt.plot(x,y2,label='BRENT',color = 'purple')
plt.title("Share Prices of Cotton and WTI over time")
plt.xlabel("Date/Time")
plt.ylabel("Share Price ($)")
plt.legend()
plt.show()
```

```python
x = data['date']
y = data['COTTON']
y1 = data['WTI']

# plt.figure(figsize=(12,5))

fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(12,5))

ax[0].plot(x,y,label="Cotton",color='green')
ax[1].plot(x,y1,label='WTI',color = 'maroon')

ax[0].set_title("Cotton Share Prices")
ax[1].set_title("WTI Share Prices")
ax[0].set_ylabel("Share Price(Cotton)")
ax[1].set_ylabel("Share Price(WTI)")
ax[0].set_xlabel("Date/Time(Cotton)")
ax[1].set_xlabel("Date/Time(WTI)")
plt.legend(ncols=2)
# ax[0].set_ylim(0,140)
plt.show()
```
